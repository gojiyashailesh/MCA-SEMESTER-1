```python
import numpy as np

#array contain only one element
arr=np.array(2)
print("Array with single element:")
print(arr)
print("\n")

#single dimensional array using list in python 
arr1 = np.array([1,2,3,4,5])
print("Single dimension array created from a list : ")
print(arr1)
print("\n")

#Multi-dimensional array(2D)
arr2 = np.array([[1,2,3],[4,5,6]])
print("multi dimensional array:")
print(arr2)
print("\n")

#3D array 
arr3 = np.array([[[1,2,3],[2,3,4],[3,4,5]]]) 
print("3D Arrays :")
print(arr3)
print("\n")

print("Dimensions of the array")
#this for checking that the array dimensons 
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
print(arr.ndim)
```
