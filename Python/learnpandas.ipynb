```python
import pandas as pd

data = pd.read_csv("C:/Users\HELLO/Downloads/titanic.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
x=data['PassengerId']
y=data["Age"]
plt.bar(x,y,label="Age",color="green")
plt.xlabel("Passenger ID")
plt.ylabel("Passenger Age")
plt.title("Passenger Information")
plt.legend()
plt.show()
```
