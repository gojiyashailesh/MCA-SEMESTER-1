```python
from array import array

'''Write a Python program to get the number of occurrences of a specified element in an
array.
Sample Output:
Original array: array(&#39;i&#39;, [1, 3, 5, 3, 7, 9, 3])
Number of occurrences of the number 3 in the said array: 3'''

def count_occurrences(arr, target):
    # Convert the array to a list to use count() method
    arr_list = list(arr)
    # Count the occurrences of the target element
    occurrences = arr_list.count(target)
    return occurrences

# Define the original array
original_array = array('i', [1, 3, 5, 3, 7, 9, 3])

# Element to count occurrences
element_to_count = 3

# Get the number of occurrences of the specified element
num_occurrences = count_occurrences(original_array, element_to_count)

# Display the result
print("Original array:", original_array)
print(f"Number of occurrences of the number {element_to_count} in the said array:", num_occurrences)
```

```python
import numpy as np

def count_occurrences(arr, elem):
    return np.count_nonzero(arr == elem)

# Sample array
my_array = np.array([1, 3, 5, 3, 7, 9, 3])
element_to_count = 3

# Count occurrences
result = count_occurrences(my_array, element_to_count)

# Display the result
print(f"Original array: array({my_array})")
print(f"Number of occurrences of the number {element_to_count} in the array: {result}")
```

```python
from array import array

def extend_array(arr, iterable):
    arr.extend(iterable)
    return arr

# Define the original array
original_array = array('i', [1, 3, 5, 7, 9])

# Iterable to append
items_to_append = [1, 3, 5, 7, 9]

# Extend the original array with items from the iterable
extended_array = extend_array(original_array, items_to_append)

# Display the result
print("Original array:", original_array)
print("Extended array:", extended_array)
```

```python
import numpy as np

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# Generating the first 20 prime numbers
primes = []
num = 2
while len(primes) < 20:
    if is_prime(num):
        primes.append(num)
    num += 1

primes_array = np.array(primes)
sum_of_primes = np.sum(primes_array)
print("Array of the first 20 prime numbers:")
print(primes_array)
print("Sum of the elements in the array:", sum_of_primes)
```

```python
import tkinter as tk
from tkinter import ttk

def sign_up():
    username = username_entry.get()
    email = email_entry.get()
    password = password_entry.get()
    
    # Add your sign-up logic here
    
    # For this example, let's print the entered details
    print(f"Username: {username}")
    print(f"Email: {email}")
    print(f"Password: {password}")

root = tk.Tk()
root.title("Social Media Sign-Up")

# Creating a frame with a glassmorphism-like effect
style = ttk.Style()
style.theme_use('clam')  # Choosing a basic theme for this example
style.configure('TFrame', background='#f0f0f0')  # Set the background color

frame = ttk.Frame(root, style='TFrame', padding=20)
frame.pack(padx=20, pady=20)

# Labels and Entries
username_label = ttk.Label(frame, text="Username:")
username_label.grid(row=0, column=0, sticky=tk.W)

username_entry = ttk.Entry(frame)
username_entry.grid(row=0, column=1)

email_label = ttk.Label(frame, text="Email:")
email_label.grid(row=1, column=0, sticky=tk.W)

email_entry = ttk.Entry(frame)
email_entry.grid(row=1, column=1)

password_label = ttk.Label(frame, text="Password:")
password_label.grid(row=2, column=0, sticky=tk.W)

password_entry = ttk.Entry(frame, show="*")
password_entry.grid(row=2, column=1)

sign_up_button = ttk.Button(frame, text="Sign Up", command=sign_up)
sign_up_button.grid(row=3, columnspan=2, pady=10)

root.mainloop()
```
